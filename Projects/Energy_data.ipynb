```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
energyDF = pd.read_csv('data/energy_dataset_.csv')
energyDF.head()
```

```python
energyDF.info()
```

```python
# 태양광만 : Type of Renwable Energy = 1
solarDF = energyDF[energyDF['Type_of_Renewable_Energy'] == 1]
solarDF.drop(['Type_of_Renewable_Energy'], axis=1, inplace=True)
solarDF.reset_index(drop=True, inplace=True)
solarDF.head()
```

Type_of_Renewable_Energy: 재생 가능 에너지원의 유형을 나타내는 숫자 코드(1: 태양열, 2: 풍력, 3: 수력, 4: 지열, 5: 바이오매스, 6: 조력, 7: 파력)  
Installed_Capacity_MW: 메가와트(MW) 단위의 설치 용량  
Energy_Production_MWh: 메가와트시(MWh)의 연간 에너지 생산량  
Energy_Consumption_MWh: 메가와트시(MWh)의 연간 에너지 소비량  
Energy_Storage_Capacity_MWh: 메가와트시(MWh) 단위의 에너지 저장 용량  
Storage_Efficiency_Percentage: 에너지 저장 시스템의 효율성(백분율)  
Grid_Integration_Level: 그리드 통합 수준을 나타내는 숫자 코드(1: 완전 통합, 2: 부분 통합, 3: 최소 통합, 4: 격리된 마이크로그리드)  
Initial_Investment_USD: USD의 초기 투자 비용  
Funding_Sources: 자금 출처를 나타내는 숫자 코드(1: 정부, 2: 민간, 3: 공공-민간 파트너십)  
Financial_Incentives_USD: USD로 된 재정적 인센티브  
GHG_Emission_Reduction_tCO2e: 온실 가스 배출량 감소(CO2 환산량(tCO2e) 톤)  
Air_Pollution_Reduction_Index: 대기 오염 감소 지수  
Jobs_Created: 생성된 작업 수
